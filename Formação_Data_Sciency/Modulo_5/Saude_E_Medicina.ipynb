```python
import pandas as pd
import matplotlib.pyplot as plt

confirmed = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv")
confirmed.rename(columns = {"Country/Region":"Country"}, inplace=True)
confirmed.query('Country.str.contains("China")')
```

```python
confirmed_by_country = confirmed.groupby("Country").sum()
plt.figure(figsize=(10,5))
confirmed_by_country.loc["China"][2:].plot()
plt.title("Casos confirmados 'China'", fontsize = 16, pad=20)
plt.ylabel("Casos", fontsize = 14)
plt.show()
```
